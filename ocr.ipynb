```python
import pytesseract 
from PIL import Image 
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# simple OCR class --> without pre-processing of the image
class BasicOCR: 
    def __init__(self):
        self.tesseract_path = '/opt/homebrew/bin/tesseract'
        pytesseract.pytesseract.tesseract_cmd = self.tesseract_path
    
    def basic_ocr(self, image_path):
        try:
            image = Image.open(image_path)
            text = pytesseract.image_to_string(image)
            return text.strip()
        except Exception as e:
            print(f"Error: {str(e)}")
            return None

ocr = BasicOCR()
print("OCR instance created successfully")
```

```python
# test BasicOCR with image (REPLACE png FILE)
image_path = 'test.png'

try:
    text = ocr.basic_ocr(image_path)
    if text:
        print("Extracted Text:")
        print("-" * 50)
        print(text)
        print("-" * 50)
    else:
        print("No text was extracted")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
# advanced OCR class that gets rid of all the noise + distractions
class AdvancedOCR:
    def __init__(self):
        self.tesseract_path = '/opt/homebrew/bin/tesseract'
        pytesseract.pytesseract.tesseract_cmd = self.tesseract_path

    def preprocess(self, image):
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)
        return thresh
    
    def process_image(self, image_path):
        try:
            image = cv2.imread(image_path)
            processed = self.preprocess(image)
            text = pytesseract.image_to_string(processed)
            return text.strip()
        except Exception as e:
            print(f"Error: {str(e)}")
            return None 

advanced_ocr = AdvancedOCR()
print("OCR instance created successfully")
```

```python
# test AdvancedOCR with image --> cleans up the image 
image_path = 'test.png'
try:
    text = advanced_ocr.process_image(image_path)
    if text:
        print("Extracted Text:")
        print("-" * 50)
        print(text)
        print("-" * 50)
    else:
        print("No text was extracted")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
